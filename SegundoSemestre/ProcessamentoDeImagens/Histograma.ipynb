# Histograma

Histograma é um grafico que representa a distribuição estatica dos niveis de cinza de imagem em termos de frequencia de ocorrencia de pixels para cada valor digital [0 a 255].

Varias medidas estatisticas podem ser obtidas a partir do histograma de uma imagem, tais como valores minimo e maximo, valor médio, variancia e desvio padrão dos niveis de cinza da imagem.

O contraste de uma imagem pode ser avaliado observando-se o seu Histograma.

```python
#Leitura da imagem
image4 = cv2.imread("/content/DianaCosmonauta.jpg")
image4 = cv2.cvtColor(image4, cv2.COLOR_BGR2RGB)

#Exibir o Histograma de todos os pixels na imagem colorida
plt.hist(image4.ravel(),bins = 256, range = [0,256]);
plt.show()
```

```python
# Mostra a imagem RGB
plt.figure(figsize=(20,10))
plt.imshow(image4)
plt.show()
```

```python
#Exibir o canal R, G, B.
color = ('r','g','b')
for i, col in enumerate(color):
  histr = cv2.calcHist([image4],[i],None,[256],[0,256])
  plt.plot(histr,color = col)
  plt.xlim([0,256])
```

```python
# converte a imagem para escala de cinza
gray_image = cv2.cvtColor(image4, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image, cmap='gray')
```

Reparem que em imagens mais escuras, o histograma correspondente vai indicar menor concentração nos valores maiores, ficando sempre proximo de zero.

# Equalização do histograma CLAHE
* CLAHE é uma variante da equalização adaptativa do histograma (AHE) que cuida da superamplificação do contraste. O CLAHE opera em
pequenas regiões da imagem, chamadas de blocos, em vez da imagem inteira. Os pixels vizinhos são então combinados usando
interpolação bilinear para remover os limites artificiais.

+ Alpha (a) corresponde ao contraste

+ Beta (b) corresponde ao brilho (brightness)

Sendo:
* alpha 1 beta 0 -> sem alteração na imagem

* O<alpha<1-> baixo contraste

* alpha > 1 —> alto contraste

* -127 < beta < +127 —> boa varidade de valores de brilho
